```python
import os
import pandas as pd
from pathlib import Path
```

```python
base_dir = "./Uilenstede"
for filename in os.listdir(base_dir):
    print(filename)
```

```python
folder = Path("./Uilenstede")

for csv_file in folder.glob("*.csv"):
    print(f"=== {csv_file.name} ===")
    try:
        df = pd.read_csv(csv_file)
        if df.empty:
            print("File is empty.\n")
            continue

        display(df.head(3))
        print("\n")

    except pd.errors.EmptyDataError:
        print("File is empty or has no columns.\n")
    except Exception as e:
        print(f"Could not read {csv_file.name}: {e}\n")
```

```python
folder_a = Path("./Uilenstede")
folder_b = Path("./Uilenstede-zoo")

# # Get CSV filenames in each folder
# files_a = {f.name for f in folder_a.glob("*.csv")}
# files_b = {f.name for f in folder_b.glob("*.csv")}

# # Intersection = files present in both folders
# common_files = sorted(files_a & files_b)

# print("Common CSV files:", common_files)

common_files = ['Accelerometer.csv', 'Compass.csv', 'Gravity.csv', 'Orientation.csv', 'Pedometer.csv']


for filename in common_files:
    print(f"\n=== {filename} ===")

    file_a = folder_a / filename
    file_b = folder_b / filename

    try:
        df_a = pd.read_csv(file_a)
        df_b = pd.read_csv(file_b)

        print("\n--- Default version (first 3 rows) ---")
        display(df_a.head(3))

        print("\n--- Sensor Zoo version (first 3 rows) ---")
        display(df_b.head(3))

    except Exception as e:
        print(f"Error reading {filename}: {e}")
```
